```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches
```

```python
x = np.random.rand(1000,2) # (10000,2) rasgele float değerlerden oluşan matris oluşturuldu.

center = np.array([[0.5,0.5]]) # Dairenin merkezi
r = 0.5 # Dairenin yarıçapı

c = ((((x - center)**2).sum(axis=1))**0.5 <= r) # Yarıçapı 0.5'e küçük eşit olan dairenin içindeki noktaları tespit ettik
                                                # c.sum() bize sayısını verir.
```

### Kenar uzunlukları 1 birim olan kare oluşturulması.

```python
N = 100
kenar0 = np.zeros(N)
kenar1 = np.ones(N)
y0 = np.linspace(0,1,N)
y1 = np.linspace(0,1,N)

kenar3 = np.zeros(N)
kenar4 = np.ones(N)
x0 = np.linspace(0,1,N)
x1 = np.linspace(0,1,N)

plt.figure(figsize=(4,4))
plt.plot(kenar0, y0, color='red')
plt.plot(kenar1, y1, color='red')
plt.plot(x0, kenar3, color='red')
plt.plot(x1, kenar4, color='red')
```

### Rassal olarak oluşturulan noktaların grafikle gösterimi

```python
plt.figure(figsize=(6,6))
plt.scatter(*x.T,c=c)

circle1 = plt.Circle((0.5, 0.5), 0.5, color='W', alpha=0.3) # Merkezi (0.5,0.5)'te olan daire oluşturuldu.
plt.gca().add_patch(circle1)

plt.plot(kenar0, y0, color='red')
plt.plot(kenar1, y1, color='red')
plt.plot(x0, kenar3, color='red')
plt.plot(x1, kenar4, color='red')

plt.xlim(-0.1,1.1)
plt.ylim(-0.1,1.1)
```

   Atılan okların yeterince fazla olması koşuluyla, bu okların kare içinde, (0.5, 0.5) merkezli ve yarıçapı da 0.5 olan dairenin içine düşme olasılığı, dairenin alanının karenin alanına oranına yakınsamalıdır.

### Oluşturulan 1000 noktanın pi sayısına yakınsadığının gösterilmesi;

```python
oran = c.sum()/1000 # Nokta sayısı yeterince fazla olmadığı için pi'ye çok yakınsamadı.
pi = oran * 4
pi
```

### Farklı sayıda oluşturulan noktaların pi sayısına yakınsadığının gösterilmesi

```python
sample = range(10000,1000000,10000)
piYakin = []
center = np.array([[0.5,0.5]])
r = 0.5
for i in sample:
    x = np.random.rand(i,2)
    
    c = ((((x - center)**2).sum(axis=1))**0.5 <= r)
    pi = (c.sum()/i)*4
    piYakin.append(pi)
```

### pi sayısının 3.14'e yakınsadığının grafik ile gösterimi

```python
plt.figure(figsize=(10,8))
plt.plot(sample,piYakin)
plt.xlabel("Nokta Sayısı")
plt.ylabel("Pi Sayısı")
plt.grid()
```
