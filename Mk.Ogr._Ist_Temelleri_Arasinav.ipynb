### Değişken grafiklerinden normal dağılıma benzeyenlere Anderson-Darling testi uygulayarak sonuçları kontrol edersek

```python
from scipy.stats import anderson
```

```python
stat, crt, sig_lvl = anderson(df['#Atış'], dist = 'norm') # "#Atış" değişkeni için örnek
alpha = 0.05
if crt[2] > alpha:
    print("#Atış", "değişkeni", 'Normal(gaussian) dağılımamış')
else:
    print("#Atış", "değişkeni", 'Normal(gaussian) dağılmış')
```

```python
df1 = df.drop(["Oyuncular", "LigSeviyesi", "LigSonuSeviye", "YeniSezonSeviyesi"],axis=1)
```

```python
df1.columns[0:6] # grafiklerden normal dağılıma yakın dağılmış olan değişkenler
```

```python
for i in df1.columns[0:6]:
    stat, crt, sig_lvl = anderson(df1[i], dist = 'norm')
    alpha = 0.05
    if crt[2] > alpha:
        print(i, "değişkeni", 'Normal(gaussian) dağılımamış')
    else:
        print(i, "değişkeni", 'Normal(gaussian) dağılmış')
```

### Anderson-Darling testine göre '#Atış', '#İsabet', '#başarılı', '#kaçanAtış', '#yakalanan', '#yürüyüş' değişkenleri normal dağılmamıştır. Fakat grafiklerden ve tanımlayıcı istatiksel bilgilere dayanarak normal dağılıma yakın bir dağılım sergilediğini söyleyebiliriz.

```python
df1.columns[6:] # grafiklerden exponential dağılıma yakın dağılmış olan değişkenler
```

```python
for i in df1.columns[6:]:
    stat, crt, sig_lvl = anderson(df1[i], dist = 'expon')
    alpha = 0.05
    if crt[2] > alpha:
        print(i, "değişkeni", 'exponential dağılımamış')
    else:
        print(i, "değişkeni", 'exponential dağılmış')
```

### Anderson-Darling testine göre '#tecrübe(yıl)', '#toplamAtış', '#toplamİsabet', '#toplambaşarılı',       '#toplamKaçanAtış', '#toplamAtış.1', '#toplamYürüyüş', '#dışarıAtış','#asist', '#hata', '#yıllıkMaaş' değişkenleri exponential dağılmamıştır. ### 6. İlgili veri seti için varyans-kovaryans matrisini hesaplayarak sonuçları yorumlayınız.

Kovaryans: İki rassal değişken arasındaki doğrusal ilişkinin değişkenliğini ölçmek için kullanılan bir kavramdır.

 a.  Pozitif olması artan bir doğrusal ilişkiyi, 
 
 b.  Sıfır civarında olması ilişkinin olmadığını, 
 
 c.  Negatif olması azalan bir doğrusal ilişkiyi gösterir.
 
 Yukarıda verilen bilgiler ışığında kovaryans matrisine yönelik yorumları kapsayan kod aşağıda sunulmuştur.

```python
kovaryans = df.cov()
kovaryans
```

```python
for i in df1.columns:
    for j in df1.columns:
        if kovaryans[i][j] < -0.99:
            print(i,"ile",j, "arasında negatif bir ilişki var",)
        elif (kovaryans[i][j] >= -0.99) & (kovaryans[i][j] <= +0.99):
            print(i, "ile", j, "arasında bir ilişki yok")
        elif (kovaryans[i][j] >= +0.99):
            print(i,"ile", j, "arasında pozitif bir ilişki var")
```

### Kovaryans matrisindeki negatif değerler değişkenler arasında ters yönlü bir ilişki olduğunu, pozitif değerler ise aynı yönde bir ilişki olduğunu gösterir.

### 11. Bir yarışma programında kırmızı yeşil ve mavi boyalı üç tane kapı gösterilmektedir. Bu kapılardan birinin arkasında büyük ödül bulunmaktadır. Yarışmacının Kırmızı kapıyı seçtiğini düşünelim. Program sunucusu hangi kapının arkasında ödül olduğunu bilmektedir. Ancak format gereği ne arkasında ödül olan kapıyı ne de yarışmacının seçtiği kapıyı açabilir. Program sunucusu Yeşil kapıyı açar ve arkasında ödül olmadığını gösterir ve yarışmacıya «İlk tercihiniz olan Kırmızı kapıyı değiştirmek ister misiniz?» diye sorar. Ödülün mavi veya kırmızı kapılar arkasında bulunma olasılıkları nedir? Monte Carlo simülasyonu kullanarak modelleyiniz.

```python
sonuclar = []
for i in range(10000): # deneyi 10000 kez tekrarlarsak sonuç bir değere yakınsayacaktır.
    kapılar = np.array([0,0,0])
    kapı_no = np.array([0,1,2]) #sırasıyla kırmızı, yeşil, mavi kapı
    kapılar[np.random.randint(0,3)] = 1 # ödül olan kapı
    secilen = np.random.randint(0,3) # yarışmacının seçtiği kapı
    acilan = np.random.choice(kapı_no[(kapı_no != secilen) & (kapılar !=1)])
    # sunucunun açtığı kapı. yarışmacının seçtiği ve ödül olan kapılar seçilemedi.
    
    sonuclar.append(secilen == kapılar.argmax()) # deneyi çok kez tekrarlayarak sonuçları listeye ekledim.
```

```python
print(np.mean(sonuclar)) # ödülün yarışmacının seçtiği kırmızı kapı arkasında olma olasılığı
print(1-np.mean(sonuclar)) # ödülün mavi kapı arkasında olma olasılığı
```

Eğer yarışmacı ilk kararında kalırsa ödülü kazanma olasılığı tek deneme için 1/3, kararını değiştirirse kazanma olasılığı 2/3'tür.
