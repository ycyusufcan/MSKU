```python
import pandas as pd
```

```python
N = 50000
atislar = np.random.randint(1, 7, size=N)
ucler = atislar == 3
uclerkum = ucler.cumsum()
atisrange = np.arange(1, N + 1)
```

```python
#zarlar = [1,3,3,1,1,2,3]
#zar3 =   [0 1 1 0 0 0 1]
#zarcum = [0 1 2 2 2 2 3]
#ozamanakadarkiatis = [1 2 3 4 5 6 7]
```

```python
plt.figure(figsize=(10,8))
plt.plot(uclerkum / atisrange, label='Deneysel Frekans')
#plt.semilogx()
plt.axhline(1/6, color='red', label='Teorik Deger')
plt.legend()
```
