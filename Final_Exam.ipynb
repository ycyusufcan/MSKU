### Değişkenlere uygulayacağımız testlerde ilk testin H0 hipotez tezimizin reddedilmesi durumunda (yani alternatif hipotezin kabul olması durumunda) 2. aşama testi uygulayacağız.
### 2. aşama testlerde Dunn testi ile tanı içindeki grupları bağımsız değişken karşısında istatistiki olarak anlamlı olup olmadığını test edeceğiz.

```python
from scipy import stats
```

## AGE

```python
k, w = list(zip(*df.groupby('Category (Tanı)')['Age'].groups.items()))
stat, p = stats.kruskal(*[df['Age'][w[i]] for i in range(len(w))])
print('Statistics=%.4f, p=%.4f' % (stat, p))
alpha = 0.05
if p > alpha:
    print('İstatistiksel olarak anlamlı bir farklılık yoktur (H0 hipotezi kabul edildi)')
else:
    print('İstatiksel olarak anlamlı bir farklılık vardır (H0 hipotezi reddedildi)')
```

#### Kruskal-Wallis testi sonucunda H0 hipotezi reddildiği için, ikinci aşamaya geçiyoruz ve Dunn testini uyguluyoruz.

```python
# pip install scikit-posthocs
```

```python
import scikit_posthocs as sp
pd.DataFrame(sp.posthoc_dunn([df['Age'][w[i]].tolist() for i in range(len(w))]).values, index=k, columns=k) < 0.005
```

#### True olarak dönen 0 ile 1, 0 ile 3, 1 ile 3 grupları ile Age değişkeninin her birinin aldığı değerler arasında istatistiksel olarak anlamlı bir farklılık vardır. (True = H0 hipotezi reddedildi demek)

### Değişkenlere normallik testlerini uygulayalım.

#### Shapiro-Wilk Normallik Testi

```python
from scipy.stats import shapiro
```

```python
degiskenler = ["Age" , "Sex", "ALB", "ALP", "ALT",  "AST", "BIL", "CHE", "CHOL", "CREA", "GGT", "PROT"]
```

```python
df["Sex"] = pd.get_dummies(df["Sex"]) # kategorik değişken olan "Sex" değişkenini sayısal değişkene çevrildi
```

```python
for i in degiskenler:
    stat, p = shapiro(df[i])
    print('stat=%.3f, p=%.6f' % (stat, p))
    if p > 0.05:
        print('Değişkenler Normal Dağılımdan Geliyor', i)
    else:
        print('Değişkenler Normal dağılmamış', i)
```

#### D'Agostino's K**2 Normallik Testi

```python
from scipy.stats import normaltest
```

```python
for i in degiskenler:
    stat, p = normaltest(df[i])
    print('stat=%.3f, p=%.3f' % (stat, p))
    if p > 0.05:
        print('Değişkenler Normal Dağılımdan Geliyor', i)
    else:
        print('Değişkenler Normal dağılmamış', i)
```

#### Anderson-Darling Normallik Testi

```python
from scipy.stats import anderson
```

```python
for i in degiskenler:
    result = anderson(df[i])
    print('stat=%.3f' % (result.statistic))
    for j in range(len(result.critical_values)):
        sl, cv = result.significance_level[j], result.critical_values[j]
        if result.statistic < cv:
            print(i, '%.1f%% seviyesinde normal dağılmış' % (sl))
        else:
            print(i, '%.1f%% seviyesinde normal dağılmamış' % (sl))
```
