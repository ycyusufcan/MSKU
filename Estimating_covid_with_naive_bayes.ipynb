```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Cleaned-Data.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.columns # data setteki featurelar  (değişkenler)
```

```python
df.drop("Country",axis=1,inplace=True) # "Country" değikenini düşürdüm.
```

```python
df1 = df.sample(1000)  # data setten 1000 uzunlukta bir örnek aldım ve yeni bir data frame'e atadım.
df1.shape
```

```python
df1.head()
```

```python
df1 = df1.reset_index(drop=True)  # data frame'in indeksini resetledim.
```

```python
df1["Covid-Positive"] = np.random.randint(0,2,1000) # data sette y (target) değişkeni (bağımlı değişken) olmadığı için
                                                    # random olarak 1000 uzunluklu bir değişken oluşturdum.
```

```python
pd.get_dummies(df1, drop_first=True) # kategorik veri içeren sütunları sayısala dönüştürdüm ve bir sütunu düştüm.
                                     # örneğin "Gender Female" otomatik olarak düştü. Çünkü "Gender Male=1", 
                                     # "Gender_Transgender=0" ise "Gender_male"ın 0 olduğu biliniyor.
```

```python
df1.head()
```

```python
X = df1.drop("Covid-Positive",axis=1)
y = df1["Covid-Positive"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=34)
# data setimi train ve test için böldüm
```

```python
from sklearn.naive_bayes import GaussianNB # modeli import ettim
```

```python
df1["Covid-Positive"].value_counts() # bağımlı değişkenin sayıları
```

```python
classifier = GaussianNB() # modelin bir instance'ını oluşturdum
```

```python
classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(y_test, classifier.predict(X_test))
# y bağımlı değişken suni (rassal olarak) üretildiği için sonuçlar da suni çıkmaktadır.
```

```python
print(classification_report(y_test, classifier.predict(X_test))) # sunilik score'ları da etkilemektedir.
```

```python
priorList = [0.4, 0.6] # y bağımlı değişkenine ağırlık atamak için liste oluşturdum.
```

```python
classifier1 = GaussianNB(priors=priorList) # oluşturduğum listeyi Naive Bayes'ın prior parametresine atayarak
                                           # ağırlıkları değiştirdim. (Tuning the hyper-parameters of an estimator.)
# Amaç: gerçekte daha fazla hastanın olduğu bilindiği bir durumda hasta olan sınıfın ağırlığı artırılarak daha doğru 
# tahmin yapmak
```

```python
classifier1.fit(X_train,y_train)
```

```python
confusion_matrix(y_test, classifier1.predict(X_test))
# y bağımlı değişkeninin hasta olan sınıfı (1) doğru tahmin sayısı artmış oldu.
```
