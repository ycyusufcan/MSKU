```python
df.drop("Country",axis=1,inplace=True) # "Country" deÄŸikenini dÃ¼ÅŸÃ¼rdÃ¼m.
```

```python
df1 = df.sample(1000)  # data setten 1000 uzunlukta bir Ã¶rnek aldÄ±m ve yeni bir data frame'e atadÄ±m.
df1.shape
```

```python
df1.head()
```

```python
df1 = df1.reset_index(drop=True)  # data frame'in indeksini resetledim.
```

```python
df1["Covid-Positive"] = np.random.randint(0,2,1000) # data sette y (target) deÄŸiÅŸkeni (baÄŸÄ±mlÄ± deÄŸiÅŸken) olmadÄ±ÄŸÄ± iÃ§in
                                                    # random olarak 1000 uzunluklu bir deÄŸiÅŸken oluÅŸturdum.
```

```python
pd.get_dummies(df1, drop_first=True) # kategorik veri iÃ§eren sÃ¼tunlarÄ± sayÄ±sala dÃ¶nÃ¼ÅŸtÃ¼rdÃ¼m ve bir sÃ¼tunu dÃ¼ÅŸtÃ¼m.
                                     # Ã¶rneÄŸin "Gender Female" otomatik olarak dÃ¼ÅŸtÃ¼. Ã‡Ã¼nkÃ¼ "Gender Male=1", 
                                     # "Gender_Transgender=0" ise "Gender_male"Ä±n 0 olduÄŸu biliniyor.
```

```python
df1.head()
```

```python
X = df1.drop("Covid-Positive",axis=1)
y = df1["Covid-Positive"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=34)
# data setimi train ve test iÃ§in bÃ¶ldÃ¼m
```

```python
from sklearn.naive_bayes import GaussianNB # modeli import ettim
```

```python
df1["Covid-Positive"].value_counts() # baÄŸÄ±mlÄ± deÄŸiÅŸkenin sayÄ±larÄ±
```

```python
classifier = GaussianNB() # modelin bir instance'Ä±nÄ± oluÅŸturdum
```

```python
classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(y_test, classifier.predict(X_test))
# y baÄŸÄ±mlÄ± deÄŸiÅŸken suni (rassal olarak) Ã¼retildiÄŸi iÃ§in sonuÃ§lar da suni Ã§Ä±kmaktadÄ±r.
```

```python
print(classification_report(y_test, classifier.predict(X_test))) # sunilik score'larÄ± da etkilemektedir.
```

```python
priorList = [0.4, 0.6] # y baÄŸÄ±mlÄ± deÄŸiÅŸkenine aÄŸÄ±rlÄ±k atamak iÃ§in liste oluÅŸturdum.
```

```python
classifier1 = GaussianNB(priors=priorList) # oluÅŸturduÄŸum listeyi Naive Bayes'Ä±n prior parametresine atayarak
                                           # aÄŸÄ±rlÄ±klarÄ± deÄŸiÅŸtirdim. (Tuning the hyper-parameters of an estimator.)
# AmaÃ§: gerÃ§ekte daha fazla hastanÄ±n olduÄŸu bilindiÄŸi bir durumda hasta olan sÄ±nÄ±fÄ±n aÄŸÄ±rlÄ±ÄŸÄ± artÄ±rÄ±larak daha doÄŸru 
# tahmin yapmak
```

```python
classifier1.fit(X_train,y_train)
```

```python
confusion_matrix(y_test, classifier1.predict(X_test))
# y baÄŸÄ±mlÄ± deÄŸiÅŸkeninin hasta olan sÄ±nÄ±fÄ± (1) doÄŸru tahmin sayÄ±sÄ± artmÄ±ÅŸ oldu.
```
