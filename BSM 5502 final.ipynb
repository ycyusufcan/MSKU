## Problem 5

```python
def pdf2(x, y):
    return 6 * x * y**2 * (0 <= x) * (x <= 1) * (0 <= y)* (y<= 1)
```

```python
a = np.linspace(0, 1, 100)
xx, yy = np.meshgrid(a, a)
zz = pdf2(xx.ravel(), yy.ravel()).reshape(xx.shape)
```

```python
plt.contourf(xx, yy, zz * (xx > yy) * (yy < xx**2))
```

```python
plt.contourf(xx, yy, zz * (np.array([xx, yy]).min(axis=0) < 0.5))
```

```python
plt.contourf(xx, yy, zz * (np.array([xx, yy]).max(axis=0) <= 0.75))
```
